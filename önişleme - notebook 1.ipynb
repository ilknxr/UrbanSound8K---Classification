```python
import cv2
import numpy as np
import pandas as pd
import tensorflow as tf
import librosa 
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/drive/MyDrive/UrbanSound8K.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
all_images = []
for i in range(8732):
  name = "/content/drive/MyDrive/spectrograms" +'/'+ str(data["classID"][i]) + '/' + data["slice_file_name"][i].replace(".wav", ".png", -4)
  a = cv2.imread(name)
  gray_a = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
  #print('Original Dimensions : ',gray_a.shape)
 
  scale_percent = 60 
  width = int(gray_a.shape[1] * scale_percent / 100)
  height = int(gray_a.shape[0] * scale_percent / 100)
  dim = (width, height)

  resized = cv2.resize(gray_a, dim, interpolation = cv2.INTER_AREA)
 
  #print('Resized Dimensions : ',resized.shape)

  image_norm = cv2.normalize(resized, None, alpha=0,beta=200, norm_type=cv2.NORM_MINMAX)
  cv2.imwrite(name, image_norm)
  #cv2_imshow(image_norm)
  cv2.waitKey(0)
  cv2.destroyAllWindows()

  all_images.append([image_norm, data["classID"][i]])

#print(np.array(all_images))

  





```

```python
images= []
images_id= []
for i in range(len(all_images)):
  images.append(all_images[i][0])
  images_id.append(all_images[i][1])

#print(images)

#print(images_id)
```

```python
X_train, X_test, y_train, y_test = train_test_split(images, images_id, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

```python
print(f"X_train: {len(X_train)}")
print(f"X_val: {len(X_val)}")
print(f"X_test: {len(X_test)}")
```

```python
df_train = pd.DataFrame({"image_X_train":X_train,"Label": y_train})
df_test = pd.DataFrame({"image_X_test ":X_test,"Label": y_test})
df_val = pd.DataFrame({"image_X_val":X_val,"Label": y_val})
```
